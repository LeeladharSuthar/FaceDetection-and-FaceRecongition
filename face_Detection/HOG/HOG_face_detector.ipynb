```python
import cv2
import dlib
import matplotlib.pyplot as plt
```

```python
def drawRectagle(image, detections):
    temp = image.copy()
    for face in detections:
        (x, y, w, h) = face.left(), face.top(), face.right(), face.bottom() #Drawing rectangle on the faces
        cv2.rectangle(temp, (x, y), (w, h), (0, 255, 0), 2)
    display(temp)
```

```python
def display(image): #function to display image
    temp = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(temp)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('./people2.jpg')
```

```python
face_detector = dlib.get_frontal_face_detector()
```

```python
detections = face_detector(image, 1)
```

```python
len(detections)
```

```python
display(image)
```

```python
drawRectagle(image, detections)
```
