```python
import cv2
import matplotlib.pyplot as plt
import dlib
```

```python
image = cv2.imread('./image-3.jpg')
```

```python
face_detector = dlib.cnn_face_detection_model_v1('./mmod_human_face_detector.dat')
```

```python
detections = face_detector(image, 2)
```

```python
def drawRectagle(image, detections):
    temp = image.copy()
    for face in detections:
        (x, y, w, h, c) = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence #Drawing rectangle on the faces
        cv2.rectangle(temp, (x, y), (w, h), (0, 255, 0), 2)
    display(temp)
```

```python
def display(image): #function to display image
    temp = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(temp)
    plt.axis('off')
    plt.show()
```

```python
drawRectagle(image, detections)
```
