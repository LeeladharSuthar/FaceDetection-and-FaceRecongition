```python
import cv2
import matplotlib.pyplot as plt
```

```python
eye_detector = cv2.CascadeClassifier("./Cascades/haarcascade_eye.xml")
```

```python
def drawRectagle(image, detections):
    temp = image.copy()
    for (x, y, w, h) in detections: #Drawing rectangle on the faces
        cv2.rectangle(temp, (x, y), (x+w, y+h), (0, 255, 0), 3)
    display(temp)
```

```python
def display(image): #function to display image
    temp = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(temp)
    plt.axis('off')
    plt.show()
    
```

```python
image = cv2.imread('./people1.jpg')
```

```python
detections = eye_detector.detectMultiScale(image, scaleFactor=1.2, maxSize=(40, 40))
print(len(detections))
drawRectagle(image, detections)
```
