```python
def display(image): #function to display image
    temp = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(temp)
    plt.axis('off')
    plt.show()
    
```

```python
face_detector = cv2.CascadeClassifier('./Cascades/haarcascade_frontalface_default.xml') #creating a classifier to detect faces
```

```python
detections = face_detector.detectMultiScale(image, scaleFactor=1.35) #running the detector on test image
```

```python
detections  #Result of the detector
```

```python
def drawRectagle(image, detections):
    temp = image.copy()
    for (x, y, w, h) in detections: #Drawing rectangle on the faces
        cv2.rectangle(temp, (x, y), (x+w, y+h), (0, 255, 0), 2)
    display(temp)
drawRectagle(image, detections)
```

```python
# minNeighbours = represents the minimum number of neighbors (rectangles) that a potential region (candidate) should have to be considered as a valid object.
# detectMultiScale also has parametes like minSize = (x, y) and maxSize=(x, y)
datections2 = face_detector.detectMultiScale(image2, scaleFactor=1.1, minNeighbors=5)
drawRectagle(image2, datections2)
```
