```python
import dlib
import cv2
```

```python
import matplotlib.pyplot as plt
def display(image): #function to display image
    temp = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(temp)
    plt.axis('off')
    plt.show()
```

```python
face_detector = dlib.get_frontal_face_detector()
points_detector = dlib.shape_predictor('../shape_predictor_68_face_landmarks.dat')
```

```python
image = cv2.imread('./images.jpg')
```

```python
faces = face_detector(image)
for face in faces:
    points = points_detector(image, face)
    for point in points.parts():
        # print(point) # 68 points
        cv2.circle(image, (point.x, point.y), 3, (0, 255, 0))
```
